# textBlob

```python
from textblob import TextBlob
import nltk
```

```python
text = "During a certain phase in his career, Kohli acquired the moniker Cheeku. One evening, Kohli availed himself of the services of a nearby salon and underwent a makeover. Upon displaying his new appearance to his colleagues, Kohli's assistant coach, Ajit Chowdhary, made a humorous comparison between Kohli's appearance and that of Chikoo,a character from the popular Indian comic book, Champak."

blob = TextBlob(text)
```

```python
# Tokenization
print("Words:",blob.words)
print("Sentences:",blob.sentences)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Parts of speech
print("Parts Tags :",blob.tags)
```

```python
# Sentiment Analysis
print("sentiment analysis :",blob.sentiment)
```

```python
# Noun phrase extraction
print("Noun phrase extraction :",blob.noun_phrases)
```

```python

# Spelling Correction
text2 = "I havv goodddddd spewwling."
blob2 = TextBlob(text2)
print("\nCorrected Spelling:", blob2.correct())

```

```python

#Lemmatization
print("\nLemmas:", [word.lemmatize() for word in blob.words])
```

```python

#N-grams
print("\nN-grams:", blob.ngrams(n=2))
```
