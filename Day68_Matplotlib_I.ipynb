```python
import matplotlib.pyplot as plt
x=["car","bike","scooty","bycycle"]
y=[50,60,70,80]
z=[10,20,40,40]
plt.title("parking value")
plt.xlabel("vehicles")
plt.ylabel("numbers")
plt.bar(x,y,width=0.2,color="r",edgecolor="blue")
plt.bar(x,z,width=0.2,color="g",edgecolor="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["car","bike","scooty","bycycle"]
y=[50,60,70,80]
z=[10,20,40,40]

width=0.2
p=np.arange(len(x))

p1=[j+width for j in p]

plt.xticks(p+width,x)

plt.title("parking value")
plt.xlabel("vehicles")
plt.ylabel("numbers")
plt.bar(p,y,width=0.2,color="r",edgecolor="blue")
plt.bar(p1,z,width=0.2,color="g",edgecolor="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["car","bike","scooty","bycycle"]
y=[50,60,70,80]
z=[10,20,40,40]

width=0.2
p=np.arange(len(x))

p1=[j+width for j in p]

plt.xticks(p+width/2,x)

plt.title("parking value")
plt.xlabel("vehicles")
plt.ylabel("numbers")
plt.bar(p,y,width=0.2,color="r",edgecolor="blue")
plt.bar(p1,z,width=0.2,color="g",edgecolor="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["car","bike","scooty","bycycle"]
y=[50,60,70,80]
z=[10,20,40,40]

width=0.2
p=np.arange(len(x))

p1=[j+width for j in p]

plt.xticks(p+width/2,x,rotation=30)

plt.title("parking value")
plt.xlabel("vehicles")
plt.ylabel("numbers")
plt.bar(p,y,width=0.2,color="r",edgecolor="blue")
plt.bar(p1,z,width=0.2,color="g",edgecolor="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["car","bike","scooty","bycycle"]
y=[50,60,70,80]
z=[10,20,40,40]

width=0.2
p=np.arange(len(x))

p1=[j+width for j in p]

plt.xticks(p+width/2,x,rotation=30)

plt.title("parking value")
plt.xlabel("vehicles")
plt.ylabel("numbers")
plt.barh(p,y,color="r",edgecolor="blue")
plt.bar(p1,z,color="g",edgecolor="blue")
plt.show()
```
