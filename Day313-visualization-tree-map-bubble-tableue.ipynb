```python
üî∑Day 313-DataScienceJourney(Tableau)
‚≠ê Topic : Tableau 
üé∏ Today , I explored :
üî•Visualization Charts in Tableau
-Tree Map
What it is:
A Tree Map displays data using rectangles.

Size = measure (e.g., Sales, Profit)

Color = another measure or category

Best used when:

Comparing values across categories & sub-categories

Seeing part-to-whole relationships

Example:
Sales by Category ‚Üí Sub-Category

How to create in Tableau:

Drag a dimension (Category) to Rows

Drag a measure (Sales) to Columns

Click Show Me ‚Üí Tree Map

Add another dimension to Color or Label

-Packed Bubble
What it is:
Uses circles (bubbles) to represent data.

Bigger bubble = larger value

Best used when:

Comparing sizes between categories

You want a visually engaging chart

No need for precise comparisons

Example:
Profit by Product or Brand

How to create in Tableau:

Drag a dimension to Detail

Drag a measure to Size

Click Show Me ‚Üí Packed Bubbles

Add color for better distinction

-Map
What it is:
Displays data based on geographic locations.

Best used when:

Analyzing data by country, state, city

Showing regional trends

Example:
Sales by State or Country

How to create in Tableau:

Drag a geographic field (Country, State) to the view

Tableau automatically creates a map

Drag a measure (Sales) to Color or Size

Use Map ‚Üí Map Layers for customization

üìä Quick Comparison
Chart Type	Best For	Visual Focus
Tree Map	Hierarchy & proportions	Rectangles
Packed Bubble	Size comparison	Circles
Map	Geographic analysis	Locations
```
