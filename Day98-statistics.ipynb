# Bernoulli Distribution 

✅ Bernoulli Distribution
The Bernoulli distribution is defined for a single trial with two possible outcomes: success (1) and failure (0). The probability mass function (PMF) of a Bernoulli random variable ( X ) can be expressed as:
[
P(X = x) =
\begin{cases}
p & \text{if } x = 1 \
1 - p & \text{if } x = 0
\end{cases}
]

where:
( p ) is the probability of success (0 ≤ ( p ) ≤ 1).
( 1 - p ) is the probability of failure.

Properties of Bernoulli Distribution
Mean (Expected Value): The mean of a Bernoulli random variable is given by: [ E(X) = p ]

Variance: The variance of a Bernoulli random variable is calculated as: [ Var(X) = p(1 - p) ]

Applications: The Bernoulli distribution is used in scenarios where there are two possible outcomes, such as coin flips, yes/no questions, and success/failure experiments.


# Binomial Distribution

✅Binomial Distribution
The Binomial distribution models the number of successes in ( n ) independent Bernoulli trials, each with the same probability of success ( p ). The PMF of a Binomial random variable ( Y ) can be expressed as:
[
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k}
]

where:
( n ) is the number of trials.
( k ) is the number of successes (0 ≤ ( k ) ≤ ( n )).
( \binom{n}{k} ) is the binomial coefficient, calculated as: [ \binom{n}{k} = \frac{n!}{k!(n - k)!} ]

Properties of Binomial Distribution
Mean (Expected Value): The mean of a Binomial random variable is given by: [ E(Y) = n \cdot p ]

Variance: The variance of a Binomial random variable is calculated as: [ Var(Y) = n \cdot p \cdot (1 - p) ]

Applications: The Binomial distribution is widely used in various fields such as quality control, clinical trials, and survey analysis, where the outcome of interest is the number of successes in a fixed number of trials.
