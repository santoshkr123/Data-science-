# Data Integrity in Data Definition Language

-Data Integrity is used to maintain the accuracy and consistency of data in a table.
-Data integrity in databases refers to the accuracy , completeness and consistency of the data stored in a database . It is the measure of the reliability and trustworthness of the data and ensures that the data in a database is protected from errors , corruption , or unauthorized changes.

Classification of data Integrity : 
1. System / Predefines integrity
2. User-defines integrity
