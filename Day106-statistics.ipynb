# Chi Square Test with python

```python
import scipy.stats as stat
import numpy as np
```

```python
# NUMBER OF HOURS STUDENTS IN A WEEKLY BASIS DAILY STUDY
# monday , tuesday,wednesday ,thursday ,friday ,saturday ,sunday

expected_data=[8,6,7,9,6,9,7]
observed_data=[7,8,6,9,9,6,7]
```

```python
chisquare_test_statistics=stat.chisquare(expected_data,observed_data)
```

```python
chisquare_test_statistics
```

```python
chisquare_test_statistics,p_value=stat.chisquare(expected_data,observed_data)
```

```python
chisquare_test_statistics
```

```python
p_value
```

```python
# find the critical value

significance_value=0.05
dof=len(expected_data)-1
print(dof)
critical_value=stat.chi2.ppf(1-significance_value,dof)
```

```python
critical_value
```

```python
if chisquare_test_statistics > critical_value:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```
