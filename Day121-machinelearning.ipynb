# Multiple Linear Regression

𝗠𝘂𝗹𝘁𝗶𝗽𝗹𝗲 𝗟𝗶𝗻𝗲𝗮𝗿 𝗥𝗲𝗴𝗿𝗲𝘀𝘀𝗶𝗼𝗻 is an extension of simple linear regression where 𝗺𝗼𝗿𝗲 𝘁𝗵𝗮𝗻 𝗼𝗻𝗲 𝗶𝗻𝗱𝗲𝗽𝗲𝗻𝗱𝗲𝗻𝘁 𝘃𝗮𝗿𝗶𝗮𝗯𝗹𝗲 is used to predict the dependent variable.


🧠 𝗖𝗼𝗻𝗰𝗲𝗽𝘁:
Equation: y=b0+b1x1+b2x2+...bnxn

y: Dependent variable
x,x2,x3,,,,,xn Independent variables
b₀: Intercept
b₁, ..., bₙ: Coefficients

🧪 𝗥𝗲𝗮𝗹-𝗟𝗶𝗳𝗲 𝗘𝘅𝗮𝗺𝗽𝗹𝗲:
Imagine predicting house prices using:
Area of the house (x₁)
Number of bedrooms (x₂)
Distance to city center (x₃)

MLR will learn how each feature affects the price and give you a formula to predict price for any house with those inputs.

# 𝗣𝗼𝗹𝘆𝗻𝗼𝗺𝗶𝗮𝗹 𝗥𝗲𝗴𝗿𝗲𝘀𝘀𝗶𝗼𝗻


 𝗪𝗵𝗮𝘁 𝗶𝘀 𝗣𝗼𝗹𝘆𝗻𝗼𝗺𝗶𝗮𝗹 𝗥𝗲𝗴𝗿𝗲𝘀𝘀𝗶𝗼𝗻?
Polynomial Regression is a type of regression where the 𝗿𝗲𝗹𝗮𝘁𝗶𝗼𝗻𝘀𝗵𝗶𝗽 𝗯𝗲𝘁𝘄𝗲𝗲𝗻 𝘁𝗵𝗲 𝗶𝗻𝗱𝗲𝗽𝗲𝗻𝗱𝗲𝗻𝘁 𝘃𝗮𝗿𝗶𝗮𝗯𝗹𝗲 𝗮𝗻𝗱 𝗱𝗲𝗽𝗲𝗻𝗱𝗲𝗻𝘁 𝘃𝗮𝗿𝗶𝗮𝗯𝗹𝗲 𝗶𝘀 𝗺𝗼𝗱𝗲𝗹𝗲𝗱 𝗮𝘀 𝗮𝗻 𝗻𝘁𝗵-𝗱𝗲𝗴𝗿𝗲𝗲 𝗽𝗼𝗹𝘆𝗻𝗼𝗺𝗶𝗮𝗹.

It is used when data shows a 𝗻𝗼𝗻-𝗹𝗶𝗻𝗲𝗮𝗿 𝗿𝗲𝗹𝗮𝘁𝗶𝗼𝗻𝘀𝗵𝗶𝗽.



🧠 𝗪𝗵𝘆 𝘂𝘀𝗲 𝗣𝗼𝗹𝘆𝗻𝗼𝗺𝗶𝗮𝗹 𝗥𝗲𝗴𝗿𝗲𝘀𝘀𝗶𝗼𝗻?
Use it when:
Your data has curvature or non-linearity.
Simple linear regression underfits the data.

📈 𝗩𝗶𝘀𝘂𝗮𝗹 𝗘𝘅𝗽𝗹𝗮𝗻𝗮𝘁𝗶𝗼𝗻:
Linear Regression fits a straight line.
Polynomial Regression fits a curve (parabola, cubic, etc.) to capture the non-linear pattern

🧪 𝗥𝗲𝗮𝗹-𝗟𝗶𝗳𝗲 𝗘𝘅𝗮𝗺𝗽𝗹𝗲:
Scenario: Predict car price based on its age.
Linear regression may assume a constant rate of depreciation.
Polynomial regression can model rapid early depreciation and slower later depreciation, capturing the real-world trend better.
