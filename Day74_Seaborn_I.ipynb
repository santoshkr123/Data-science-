# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

var=[1,2,3,4,5,6,7]
var1=[7,6,5,4,3,2,1]

plt.plot(var,var1)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=[1,2,3,4,5,6,7]
var1=[7,6,5,4,3,2,1]

df=pd.DataFrame({"var":var,"var1":var1})
df

sns.lineplot(x='var',y='var1',data=df)



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1,bins=[170,180,190,200,210,230])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1,bins=[170,180,190,200,210,230],kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1,bins=[170,180,190,200,210,230],kde=True,rug=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1,bins=[170,180,190,200,210,230],kde=True,rug=True,color="r")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df1=sns.load_dataset("Penguins")
sns.displot(x="flipper_length_mm",data=df1,color="r",log_scale=True)
plt.show()
```
