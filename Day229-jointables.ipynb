# JOIN TABLES 

```python
- Cross Join
-Self Join
-Join on more than one column
-Join on multiple table
-Filtering Columns
-Filtering Rows

ðŸ”¥Cross Join : The Cross Join keyword returns all records from both tables(table1 and table2)
select column_name(s)
from table1
cross join table2;

ðŸ”¥Self Join : Use with WHERE clause or where condition
A self join is a regular join ,but the table is joined with itself.
select column-name(S)
from table1,table2
where condition;
```

```python
select * from customers
cross join orders;

select *
from customers , orders
where amount > 500;

select customers.customer_id,customers.first_name ,orders.amount
from customers
left join orders
on customers.customer_id=orders.customer
where amount > 500;
```
