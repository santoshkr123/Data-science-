```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
z=[12,13,14,15,16]

plt.title("scatter plot", fontsize=20)

plt.xlabel("month" , fontsize=20)
plt.ylabel("number" , fontsize=20)

size=[300,200,500,250,120]

c=[20,30,40,50,60]

plt.scatter(x,y, c=c , s=size , cmap="jet")
t=plt.colorbar()
t.set_label("colorbar" , fontsize=20)


plt.scatter(x,z, c=c , s=size , cmap="summer")
t=plt.colorbar()
t.set_label("colorbar" , fontsize=20)
plt.show()
```
