# GROUP BY IN SQL 

```python
ðŸ”¥ Grouping(Group by )
-Fundamental of SQL grouping
-Grouping average function
-Group by multiple columns
-Grouping descending and ascending orders
-Group by use with limit clause
-Having clause used with the group by 

ðŸ”¥Group by SQL Statement 
-The Group By Statement group by rows that have the some values into summary rows
-The Group by statement is often used with aggregate functions(count(),max(),min(),sum(),avg()) to group theresult-set by one or more columns.
```

```python
select brand_name , count(*) as 'Total_Phones' ,
round(avg(price)) as 'AVG_Price' ,
max(rating) as 'MAX_Rating' ,
round(avg(battery_capacity)) as 'Battery_Cap',
min(screen_size) as 'Screen_size'
from smartphones_cleaned_v6
group by brand_name 
order by Total_Phones desc limit 5;


select has_5G,
avg(price) as 'AVG_Price',
avg(rating)as 'AVG_Rating'
from smartphones_cleaned_v6
group by has_5g;

select has_5G, has_ir_blaster,
avg(price) as 'AVG_Price',
avg(rating)as 'AVG_Rating'
from smartphones_cleaned_v6
group by has_5g , has_ir_blaster;

select brand_name  , processor_brand,
count(*) as 'No_Of_Phones',
avg(battery_capacity) as 'batter_cap'
from smartphones_cleaned_v6
group by brand_name , processor_brand;

select brand_name , count(*) as 'total_phones'
from smartphones_cleaned_v6
group by brand_name 
order by total_phones asc limit 5
```
