# Violin plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset("tips")
df
```

```python
sns.violinplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time",linewidth=3)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time",linewidth=1,palette="rainbow_r")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time",linewidth=1,palette="magma_r")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df  ,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,split=True,hue="sex")
plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data=df,split=True,hue="sex")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df  ,order=["Dinner","Lunch"],inner="quartz")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df  ,order=["Dinner","Lunch"],inner="box")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df  ,order=["Dinner","Lunch"],inner="point")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df  ,order=["Dinner","Lunch"],inner="stick")
plt.show()
```

```python
sns.violinplot(x=df["total_bill"])
plt.show()
```

```python
sns.violinplot(y=df["total_bill"])
plt.show()
```

# pairplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue="sex")
plt.show()
```

```python
sns.pairplot(df,hue="sex",vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(df,hue="sex",vars=["total_bill","tip"],hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(df,hue="sex",vars=["total_bill","tip"],hue_order=["Female","Male"],palette="rainbow_r")
plt.show()
```

```python
sns.pairplot(df,hue="sex",x_vars=["total_bill","tip"],hue_order=["Female","Male"],palette="rainbow_r")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],palette="rainbow_r")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],
             palette="rainbow_r",kind="reg")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],
             palette="rainbow_r",kind="hist")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],
             palette="rainbow_r",kind="scatter")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],
             palette="rainbow_r",kind="kde")
plt.show()
```

```python
sns.pairplot(df,hue="sex",y_vars=["total_bill","tip"],hue_order=["Female","Male"],
             palette="rainbow_r",diag_kind="hist")
plt.show()
```

# boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",color="r")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",order=["Fri","Thur","Sat","Sun"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",order=["Fri","Thur","Sat","Sun"],
            showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",order=["Fri","Thur","Sat","Sun"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},palette="magma_r")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=df,
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},palette="magma_r")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=df,
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},palette="magma_r",orient="h")
plt.show()
```

```python
sns.boxplot(y=df["total_bill"])
```
