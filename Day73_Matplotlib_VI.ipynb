# Step Plot

# Fill Between Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,color="r")
plt.fill_between(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,color="r")
plt.fill_between(x,y)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,color="r")
plt.fill_between(x=[ 2,4],y1=2,y2=4)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y ,where=(x>=2)& (y<=4))
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y ,where=(x>=3)& (y<=4.5))
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y ,where=(x>=3)& (y<=4.5),alpha=0.4)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Sub plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

x1=[1,2,3,4,5]
y1=[4,5,8,9,10]
plt.subplot(2,2,2)
plt.step(x1,y1)

x3=[5,8,9,7,15]
plt.subplot(2,2,3)
plt.pie(x3)

x5=["a","b","c","d","e"]
x6=[5,6,8,9,4]
plt.subplot(2,2,4)
plt.bar(x5,x6)
plt.show()
```

# Text

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.title("Graph")
plt.text(2,8,"Text",fontsize=20,style="italic")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.title("Graph")
plt.text(2,8,"Text",fontsize=20,style="italic",bbox={"facecolor":"red"})
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.title("Graph")
plt.text(2,8,"Text",fontsize=20,style="italic",bbox={"facecolor":"red"})
plt.annotate("student",xy=(1,6),xytext=(4,8),arrowprops=dict(facecolor="g",shrink=100))
plt.show()
```
