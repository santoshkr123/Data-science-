```python
y={"fontsize":20,"color":"blue"}
sns.heatmap(var,vmin=0,vmax=10,annot=v,fmt="s",annot_kws=y,linewidth=4,linecolor="y",
            cbar=False,xticklabels=False)
plt.show()
```

```python
y={"fontsize":20,"color":"blue"}
sns.heatmap(var,vmin=0,vmax=10,annot=v,fmt="s",annot_kws=y,linewidth=4,linecolor="y",
            cbar=False,yticklabels=False)
plt.show()
```

```python
y={"fontsize":20,"color":"blue"}
v=sns.heatmap(var,vmin=0,vmax=10,annot=v,fmt="s",annot_kws=y,linewidth=4,linecolor="y",
            cbar=False,yticklabels=False)

v.set(xlabel="Python",ylabel="code")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


var = pd.DataFrame(np.random.randint(0, 11, size=(5, 5)))

y = {"fontsize": 20, "color": "blue"}


v = sns.heatmap(var, vmin=0, vmax=10, annot=True, fmt="d", annot_kws=y,
                linewidth=4, linecolor="y", cbar=False, yticklabels=False)

v.set(xlabel="Python", ylabel="code")

plt.show()
```

# Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.load_dataset("tips")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data1=sns.load_dataset("tips")

sns.barplot(x="sex",y="tip",data=data1)
plt.show()
```

```python
sns.countplot(x="sex",data=data1)
plt.show()
```

```python
sns.countplot(x="sex",data=data1,hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=data1,hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=data1,hue="smoker",palette="rainbow_r")
plt.show()
```

```python
sns.countplot(x="sex",data=data1,hue="smoker",palette="rainbow_r")
plt.show()
```
