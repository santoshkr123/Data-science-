# Box-Whisper Plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,180]
plt.boxplot(x,flierprops=dict(markeredgecolor="y"))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,180]
x1=[10,20,30,40,50,60,80,90,200]
y=[x,x1]
plt.boxplot(y,flierprops=dict(markeredgecolor="y"))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,180]
x1=[10,20,30,40,50,60,80,90,200]
x2=[10,20,30,40,50,60,80,90,200,300]
y=[x,x1,x2]
plt.boxplot(y,flierprops=dict(markeredgecolor="y"))
plt.show()
```

# Stack plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[6,7,8,9,10]
plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[11,13,50,18,20]
plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[11,13,50,18,20]
area1=[15,19,22,63,88]
area2=[55,36,89,77,96]
plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[11,13,50,18,20]
area2=[15,19,22,63,88]
area3=[55,36,89,77,96]
l=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[11,13,50,18,20]
area2=[15,19,22,63,88]
area3=[55,36,89,77,96]
l=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[11,13,50,18,20]
area2=[15,19,22,63,88]
area3=[55,36,89,77,96]
l=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","magenta"])
plt.title("graph")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[11,13,50,18,20]
area2=[15,19,22,63,88]
area3=[55,36,89,77,96]
l=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","magenta"])
plt.title("graph")
plt.xlabel("record")
plt.ylabel("numbers")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[11,13,50,18,20]
area2=[15,19,22,63,88]
area3=[55,36,89,77,96]
l=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","magenta"])
plt.title("graph",fontsize=20)
plt.xlabel("record",fontsize=20)
plt.ylabel("numbers",fontsize=20)
plt.legend(loc=3)
plt.show()
```
