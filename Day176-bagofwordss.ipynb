# Co-occurance Matrix 

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text="""data science is amazing
machine learning is a part of data science 
deep learning and machine learning are useful
data science and deep learning go hand in hand"""


words=text.lower().split()#Lowercase and split into words
vocabulary=sorted(list(set(words)))# Get unique words and sort them
word_to_index={word:i for i,word in enumerate(vocabulary)}
window_size=1
```

```python
vocabulary
```

```python
word_to_index
```

```python
matrix=np.zeros((len(vocabulary),len(vocabulary)),dtype=int)


for i in range(len(words)):
    current_word=words[i]
    for j in range(max(0,i-window_size),min(len(words),i+window_size+1)): #sliding window
        if i!=j:
            neighbour_word=words[j]
            matrix[word_to_index[current_word],word_to_index[neighbour_word]]+=1
```

```python
df =pd.DataFrame(matrix,index=vocabulary,columns=vocabulary)
```

# word2vec

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
corpus=[
    "The cat sits on the mat",
    "The dog plays in the garden",
    "Dogs and cats are god pets",
    "The garden has beautiful flowers "
]
```

```python
tokenized_corpus=[word_tokenize(sentence.lower()) for sentence in corpus]
```

```python
# Train Word2Vec model
model=Word2Vec(sentences=tokenized_corpus,vector_size=100,window=5,min_count=1,workers=4)
```

```python
# Get word vector for 'cat'
print(model.wv['cat'])
```

```python
# find similar words 
print(model.wv.most_similar('dog'))
```

```python
print(model.wv.most_similar('dog'))
```

# Doc2vec


```python
import gensim
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
import nltk
from nltk.tokenize import word_tokenize
```

```python
documents=[
    "Data Science is an interdisciplinary field.",
    "Machine learning models improve predictions.",
    "Artificial Intelligence is evolving rapidly.",
    "Python is widely used for data analytics"
]
```

```python
tagged_data=[TaggedDocument(words=word_tokenize(doc.lower()),tags=[str(i)]) for i ,doc in enumerate(documents)]
tagged_data
```

```python
# Train Doc2Vec model
model=Doc2Vec(vector_size=50,
             window=2,
             min_count=1,
             workers=4,
             epochs=100)
model.build_vocab(tagged_data)
```
