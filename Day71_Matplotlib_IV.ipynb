# Pie plot in matplotlib


```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["maths","english","sanskrit","SST"]
plt.pie(x,labels=y)
plt.show()
```

# stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="+")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*",bottom=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*",bottom=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="magenta")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="magenta",label="code")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="magenta",label="code",orientation="horizontal")
plt.legend()
plt.show()
```
