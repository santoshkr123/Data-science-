📊 Probability Concepts in Statistics
1. Probability Distribution Function (PDF - general term)
🔁 A general term for any function that describes the probability distribution of a random variable.

✅ Can refer to:

Probability Mass Function (PMF) for discrete variables

Probability Density Function (pdf) for continuous variables

2. Probability Mass Function (PMF)
🔸 Used for discrete random variables (e.g., number of heads when flipping a coin).

🔢 It gives the probability of each possible outcome.

📌 Example: P(X = 2) = 0.25 means the probability of getting value 2 is 0.25.

🧠 Think: Counting distinct outcomes like dice rolls, coin tosses.

3. Probability Density Function (pdf)
🔸 Used for continuous random variables (e.g., height, weight).

📈 It gives the relative likelihood of a value falling in an interval.

⛔ You cannot get the probability of a specific value (e.g., P(X = 3)) — always ranges:
📌 Example: P(2 < X < 4)

🧠 Think: Areas under the curve using integration.

4. Cumulative Distribution Function (CDF)
📊 Applies to both discrete and continuous variables.

🔼 CDF at a value x gives the probability that X is less than or equal to x:

📌 F(x) = P(X ≤ x)

✅ It's a non-decreasing function that starts from 0 and goes to 1.

🧠 Think: Accumulated probabilities.
