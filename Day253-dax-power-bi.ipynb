* Calculated Column and Measure 
Calculated Column - Expands Table By creating new column.
-stored along with table , consumer memory

Measure -
Summarized data into a single value
-calculates at runtime and stored temporarily
-rich analytical capabilities

*when to use calculated column ?
-when you want to your calculations to be done on a row level basis then we enter Dax into a calculated column

*when to use measure ?
-When you want to do the calculation on an aggregated level and don't particularly want to store information on the row level
-In comparison to columns,measures don;t create physical values in your table , hence do not increase file size. That; only why must be used when you don;t want to spend extra space .

*Row Context :Row context does calculation for each row , with the values within that row.

*Filter Context : Filter Context refers to the filters that are applied to the data when evaluating a Dax expression .

* Calculate function : It 's used to modify the filter context and evaluate expressions in a different context
