🔢 𝟭. 𝗥-𝗦𝗾𝘂𝗮𝗿𝗲𝗱 (𝗥𝟮) – 𝗖𝗼𝗲𝗳𝗳𝗶𝗰𝗶𝗲𝗻𝘁 𝗼𝗳 𝗗𝗲𝘁𝗲𝗿𝗺𝗶𝗻𝗮𝘁𝗶𝗼𝗻
𝗣𝘂𝗿𝗽𝗼𝘀𝗲: Measures how well the regression model explains the variance in the target variable.
𝗥𝗮𝗻𝗴𝗲: 0 to 1
𝗜𝗻𝘁𝗲𝗿𝗽𝗿𝗲𝘁𝗮𝘁𝗶𝗼𝗻:
R² = 0 → Model explains none of the variability.
R² = 1 → Model explains all the variability.
𝗘𝘅𝗮𝗺𝗽𝗹𝗲: An R² of 0.85 means the model explains 85% of the variance in the dependent variable.
🧠 Use R² when you want to understand how well your model fits the data.

📉 𝟮. 𝗔𝗱𝗷𝘂𝘀𝘁𝗲𝗱 𝗥-𝗦𝗾𝘂𝗮𝗿𝗲𝗱
𝗣𝘂𝗿𝗽𝗼𝘀𝗲: A modified version of R² that adjusts for the number of predictors.
𝗪𝗵𝘆 𝗶𝘁'𝘀 𝗻𝗲𝗲𝗱𝗲𝗱: Adding more features can artificially inflate R²; Adjusted R² penalizes this.
🧠 Use Adjusted R² when comparing models with different numbers of features.

𝗠𝗲𝗮𝗻 𝗔𝗯𝘀𝗼𝗹𝘂𝘁𝗲 𝗘𝗿𝗿𝗼𝗿 (𝗠𝗔𝗘)
𝗣𝘂𝗿𝗽𝗼𝘀𝗲: Measures the average absolute difference between predicted and actual values.
𝗜𝗻𝘁𝗲𝗿𝗽𝗿𝗲𝘁𝗮𝘁𝗶𝗼𝗻: Easy to understand; less sensitive to outliers.

 𝗠𝗲𝗮𝗻 𝗦𝗾𝘂𝗮𝗿𝗲𝗱 𝗘𝗿𝗿𝗼𝗿 (𝗠𝗦𝗘)
𝗣𝘂𝗿𝗽𝗼𝘀𝗲: Measures the average of the squared differences between predicted and actual values.
𝗜𝗻𝘁𝗲𝗿𝗽𝗿𝗲𝘁𝗮𝘁𝗶𝗼𝗻: Penalizes larger errors more than smaller ones.
🧠 Useful when large errors are particularly undesirable.

📐𝗥𝗼𝗼𝘁 𝗠𝗲𝗮𝗻 𝗦𝗾𝘂𝗮𝗿𝗲𝗱 𝗘𝗿𝗿𝗼𝗿 (𝗥𝗠𝗦𝗘)
𝗣𝘂𝗿𝗽𝗼𝘀𝗲: Square root of MSE; gives error in the same units as the target variable.
𝗜𝗻𝘁𝗲𝗿𝗽𝗿𝗲𝘁𝗮𝘁𝗶𝗼𝗻: Easier to interpret than MSE since it's in the original units.

⚖️ 𝗪𝗵𝗶𝗰𝗵 𝗠𝗲𝘁𝗿𝗶𝗰 𝗦𝗵𝗼𝘂𝗹𝗱 𝗬𝗼𝘂 𝗨𝘀𝗲?
For 𝗺𝗼𝗱𝗲𝗹 𝗳𝗶𝘁: R² and Adjusted R²
For 𝗲𝗿𝗿𝗼𝗿 𝘀𝗶𝘇𝗲: MAE (robust), MSE (sensitive), RMSE (interpretable)
