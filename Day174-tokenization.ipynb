```python
import nltk
```

```python
corpus="""Hello My name is Soahib Khan we are learning NLP.
Plz watch the till the end.
This is going to be amazing Plz watch NLP completely for better response."""
```

```python
corpus
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences=sent_tokenize(corpus) #converting paragraph into sentences
```

```python
sentences
```

```python
type(sentences)
```

```python
for sentence in sentences:
    print(sentence)
```

```python
sentences[0]
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in sentences:
    print(sentence)
    print()
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus) #seperating 's in different word'
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
TreebankWordTokenizer().tokenize(corpus)
#here is the full stop is cont considered as a separete word except at the end of corpus
```
