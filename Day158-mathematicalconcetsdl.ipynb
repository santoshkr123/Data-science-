# Mathemetical Concept in deep learning

1. Vectors :

object {magnitude , direction}



2. Differentiation :

f(x) - Rate of change of a function with respect to its input =>deltax



3. Partial Differentiation :

f(x,y) = df / dx{ y-constant}      , df / dy {-x constant}



4 . Gradient  :

Partial derivative ->vector format

deltaf=[df / dx        df / dy]T



5.  Maxima and Minima  :

a .Univariate :

y=f(x)

step1 f'(x)

step2 critical points f'(x)=0  ,a,b=criticalmpoints

step3 f''(x)

case 1 f''(x) > 0 MINIMA

case 2 f''(x) < 0 MAXIMA

case 2 f''(x) =0 Point of Inflection



b) Bivariate  

y=f(x,y)

step 1 df / dx           df / dy 

step 2 df/dx=0    df/dy=0

step 3 double derivate

fxx    fxy    fyy

r        s        t

case1 rf-s*s > 0(max,min)

case2 rf-s*s=0 not valid 

case3 rf-s*s<0 ->Neither minima nor maxima
