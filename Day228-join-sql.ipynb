# JOIN _SQL

```python
What is SQL Join
-Types of Join
-INNER Join
-LEFT Join
-RIGHT Join
-Full(Outer Join)

 ðŸ“Œ JOIN - A JOIN clause is used to combine rows from two or more tables, based on a related column between them

ðŸ“ŒINNER JOIN - Returns records that have matching values in both tables.

ðŸ“ŒLEFT JOIN - Returns all records from the left table , and the matched records from the right table.

ðŸ“ŒRIGHT(Outer) JOIN - Returns all records from the right table and the matched records from the left table.

ðŸ“ŒFULL(Outer) JOIN -Returns all records when there is a match in either left or right table
```

```python

select customers.customer_id , customers.first_name , orders.amount
 from customers
inner join orders
on customers.customer_id=orders.customer;


select customers.customer_id , customers.first_name , orders.amount
 from customers
left join orders
on customers.customer_id=orders.customer;


select customers.customer_id , customers.first_name , orders.amount
 from customers
right join orders
on customers.customer_id=orders.customer;


select * from customers
left join orders
on customers.customer_id = orders.customer
UNION
select * from customers
right join orders
on customers.customer_id = orders.customer

```
