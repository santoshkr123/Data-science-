# Cardianlty of Relationship

In database management ,cardinality plays an important role . Here cardinality represent the number of times of an entity
set participated in a relationship set . Or we can say that the cardinalty of a relationship is the number of tupes(rows) in a relationship

Types of cardinality in between table are :
one-to-one
one-tomany
many-to-many
many-to-many

one-to-one(1:1)-when one entity in each entity set takes part at most once in the relationship , the cardinality is one to one

one-to-many(1:N) -If entities in the first entity set take part in the relationship set at most once and entites and entites in the second entity setn
part many times(at least twice),the cardinality is said to be one to many

many-to-one(N:1) : if entites in the first entity set take part in the relationship set ,many times(at least twice) while 
entites in the second entity set the take part at most once,  the cardinality is said to be many to one

many to many (N:N): The cardinality is said to be many to many if entites in both the entity sets take part many times (at least twice) 
in the relationship
