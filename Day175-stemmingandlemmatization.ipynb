```python
from nltk.stem import PorterStemmer
```

```python
Stemmer=PorterStemmer()
```

```python
Stemmer.stem("running")
```

```python
Stemmer.stem("eating")
```

```python
words=["eating","eats","finally","eater","finalized","programmer","programming","eaten","programs","writing","sportingly",
      "fairly"]
```

```python
for word in words:
    print(word+'>>>>>>>'+Stemmer.stem(word))
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowStemmer=SnowballStemmer("english")
```

```python
snowStemmer.stem("programmer")
```

```python
for word in words:
    print(word+'======>'+snowStemmer.stem(word))
```

```python
from nltk.stem import RegexpStemmer
```

```python
regexStemmer=RegexpStemmer('ing$|es$|s$',min=4)
```

```python
regexStemmer.stem("goes")
```

```python
for word in words:
    print(word+'======>'+regexStemmer.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('history',pos='r')
```

```python
for word in words:
    print(word+'======>'+lemmatizer.lemmatize(word,'v'))
```
