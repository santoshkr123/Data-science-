# catplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset("tips")
df
```

```python
sns.catplot(x="tip",y="size",data=df)
```

```python

df1=sns.load_dataset("tips").head(20)
df1
```

```python

sns.catplot(x="tip",y="size",data=df1)
plt.xticks(rotation="vertical")
plt.show()
```

```python

sns.catplot(x="tip",y="size",data=df1,hue="sex")
plt.xticks(rotation="vertical")
plt.show()
```

```python

sns.catplot(x="tip",y="size",data=df1,hue="sex",height=10,kind="bar")
plt.xticks(rotation="vertical")
plt.show()
```

```python

sns.catplot(x="time",y="total_bill",data=df1,hue="sex",height=10,kind="box")
plt.xticks(rotation="vertical")
plt.show()
```

```python

sns.catplot(x="tip",y="size",data=df1,hue="sex",height=10,kind="point")
plt.xticks(rotation="vertical")
plt.show()
```

# Styling plot in seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset("tips")
df
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.set_context("poster")
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
sns.despine()
```

# Facet Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df2=sns.load_dataset("tips").head(20)
df2
```

```python
fg = sns.FacetGrid(df2, col="sex",height=5,hue="day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
 fg = sns.FacetGrid(df2, col="sex",height=5,hue="day")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```
