# Constrains in MySQL

```python
Auto-increament   : Auto-increment allows a unique number to be generated automatically when a new record is inserted into a table.
Check - The CHECK constraint is used to limit the value range that can be placed in a column.
Default-The DEFAULT constraint is used to set a default value for a column.
Forenign Key -A FOREIGN KEY is a field (or collection of fields) in one table, that refers to the PRIMARY KEY in another table.
```
