# YOLO

*YOLO (You Only Look Once ): ek object detection algorithm hai jo deep learning mein istemel hota hai, Yeh algorithm bahut tez aur efficient hai kyunki yeh hi forward pass mein object detection ko kar leta hai.
Objective :
1. Single shot Detection : YOLO ek hi forward pass main object detection, krta hai matlab ek hi baar min poora image ko scan karta hai.

2. Bounding Box Prediction : YOLO bounding boxes ko seedha predict karta hai ,har grid cell mein multiple bounding boxes ko predict karta hai,

3. Anchor Boxes : YOLO ke anchor boxes ka istemel hota hai,alag-alag aur aspect size our aspect ratio ke objects ko detect karne ke liye.

4.Grid Cells : YOLO image ko grid cells mein divide karta hai aur cell mein object detection karta hai.

5. Non-max Suppression : Duplicate detection ko hatane ke liye YOLO non-max suppression ka istemel karta hai

*Architecture :
YOLOv3 mein ek deep neural network (Darknet-53) ka istemel hota hai features ko extract karne ke liye . Yeh multiple detection layers aur fixed-size tensor ko istemel karta hai bounding box coordinates , objectives scores ,aur class probabilities ke liye 

*Training Program : YOLO ke liye labeled training data ki zarurat hoti hai jisme images aur corresponding bounding box annotations hote hai, YOLO training ke duration loss function ko istemel karta hai. backpropagation ke zariye weights ko update kiya jota hai training ke duaran.

*Advantages : 
->YOLO bahut tez hai aur real time object detection ke liye istemel kiya ja skata hai.
->YOLO ek hi forward pass mein object detection karta hai , isse efficient aur accuarte hai .
 Example : YOLO ka autonomons driving , surveillance systems,aur medical imaging mein istemel kiya jata hai.

*Generative Adversarial Network(GAN) : ek deep learning model hai jo do neural networks ko ek saath train karta hai .Ye model images , audio ,video , ya anya data ko generate karta ke liye istemel hota hai 

Objective :
1. Generator : Ye network fake data generate karta hai , Iska kaam hai real data ke jaisi images ya samples banana ka.

2. Discriminator : Ye network real aur fake data ko pechanne ka kaam karta hai . Ye samjhta hai ki data real hai ya fake.

Advantages : 
GAN ka istemel high-quality fake data generate karne ke liye kiya ja sakte hai , jo ki image generation , video synthesis aur artistic creation 

Example :
GANs ko istemel karke realistic images , paintings aur human-like faces generate kiye ja skta hai .
