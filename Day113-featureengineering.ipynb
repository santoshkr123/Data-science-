```python
x=np.array([1,2,3,4,5])
y=np.array([7,8,9,10,11])
```

# Cubic Interpolation 

```python
x=np.array([1,2,4,8,9,10])
y=np.array([12,45,66,88,96,110])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```
