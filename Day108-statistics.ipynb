# 📊 ANOVA (Analysis of Variance) – Statistics

ANOVA is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different.

🔍 Why Use ANOVA?
Instead of performing multiple t-tests (which increases error), ANOVA checks all groups in a single test.

🧪 Example:
You're testing three teaching methods.
You collect scores from students in each group.
ANOVA tells you whether the difference in average scores is statistically significant.

🧮 Key Concept:
Null Hypothesis (H₀): All group means are equal

Alternative Hypothesis (H₁): At least one group mean is different
