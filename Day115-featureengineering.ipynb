# Data Encoding -One Hot Encoder

```python
import pandas as pd
```

```python
df=pd.DataFrame({'color':['red','blue','green','yellow','pink','orange']})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']])
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded=encoder.fit_transform(df[['color']])
```

```python
import pandas as pd
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
pd.concat([df,encoded_df],axis=1)
```

# Ordinal Featuring

```python
import pandas as pd
```

```python
df=pd.DataFrame({'size':['small','medium','large','small','medium']})
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder
```

```python
encoded=encoder.fit_transform(df[['size']])
```

```python
encoded
```

```python
encoded_df=pd.DataFrame(encoded)
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```

# target Guided Ordinal Encoding

```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'city':['NewYork','London','Paris','Tokyo','NewYork','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
#calculate the mean price for ech city
df.groupby('city')['price'].mean()
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
#replace each city with its mean price
df['city_encoded']=df['city'].map(mean_price)
```
