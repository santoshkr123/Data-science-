# DataBase Keys

what are the keys in DBMS ?
A key in DBMS is an attribute or  a set of attributes that help to uniquely identify a tuple 
(or new) in a relation(table)

why do we need keys ?
To identify a row and keys are also used to estabish relationship between the different 
tables and columns of a relational database
Indivisuals values in a key are called key values

# Types of keys

Super key  : A super key is a combination of columns that uniquely identifies any row within a 
relational dataabse management system(RDBMS)table

Candidate key  : A candidate key is a minimal super ke , meaning it has no redundant attributes,
In other,words it's the smallest set of attributes that can be used to uniquely identify a type(row)
in the table

Primary Key :  A primary ley is a unique identifier for each tuple in a table .
There can only be one primary key and it can't contain null values.
-No Repeat / Duplicate , UNIQUE values , and can't contain NULL values
-Prefer Numeric ,Small Length , Value not changes Always 
                                        

Alternate Keys : An Alternate keys is a candidate key that is not used as the primary key

Composite Key  : A composite key is a primary key  that is made up of two or more attributes 
key . Composite key are used when a single attributes is not sufficient to uniquelly
identify a tuple in a table

Suurogate Key  : A surrogate key is defined as a unique identifier for some record or object
in a table . It is similar to a primarly key , but with a significant difference .It 
is not derived from the table data the object generates this key itself

Foreign Key : A foreign key is a primary key from one table that is used to estabalish 
connection a relationship with another table
