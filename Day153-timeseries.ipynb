# Time Series

*Time Series : Time series data consists of observations collected at regular time intervals, such as daily stock prices, monthly sales figures, or yearly climate data. *Key components of time series data include:
Trend-Upward ,Horizontal , Downward
Season - Frequently repetition like weekly,monthly,Dya,Hour
Cycle : Over the long time -Stock price ,election 
Noise - Randomness , Residual ,WhiteNoise like Pandemic , Speech ,Report

*Interpolation -To predict the value in the given Range
*Extrapolation -To predict the value outside the given range

*Why should we not use Linear regression ?
-Interpolation-Extrapolation
-Outlier we should not use
-It having Complex data not linear
-Since in the series we have time dependency
