📌 What is Inferential Statistics?
It’s a branch of statistics used to draw conclusions and make predictions about a population based on a sample of data.

💡 Goal: Use sample data to infer or estimate characteristics of the larger population.

🔍 Example:
Survey 100 people to estimate the average salary of a city — you don’t ask everyone, just enough to make an informed guess.

📐 Estimation
🔹 Point Estimate:
A single value used to estimate a population parameter.
🧾 Example: Sample mean = 70 kg → Estimating population mean weight.

🔹 Interval Estimate:
A range within which the parameter likely falls, with a confidence level.
🧾 Example: "The average weight is 70 ± 5 kg with 95% confidence."

🧪 Hypothesis Testing
It’s a statistical method to test an assumption (hypothesis) about a population based on sample data.

🔄 Steps:
Null Hypothesis (H₀): No effect or no difference

Alternative Hypothesis (H₁): Claim you're testing

Perform test: (t-test, z-test, ANOVA, chi-square, etc.)

Make a decision: Reject or fail to reject H₀

🧠 It helps determine if sample results are statistically significant.

🔍 Real-Life Example:
You claim a new drug reduces recovery time.

H₀: No improvement

H₁: Drug improves recovery

Statistical test → Reject H₀ if evidence supports H₁.

📉 p-value: What is it?
It shows the probability of observing your data (or more extreme) if H₀ were true.

✅ Small p-value (≤ 0.05) → Strong evidence → Reject H₀
❌ Large p-value (> 0.05) → Weak evidence → Fail to reject H₀

🧾 Example:
Testing a new teaching method:

p = 0.02 → Since 0.02 < 0.05, you reject H₀ → The method works!
