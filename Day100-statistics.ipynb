# Uniform Distribution: Continuous and Discrete

# Continuous Uniform Distribution

A continuous uniform distribution describes a scenario where all outcomes in a given interval are equally likely. This type of distribution is defined over a continuous range of values, typically represented as ( U(a, b) ), where ( a ) and ( b ) are the minimum and maximum values, respectively.

Properties of Continuous Uniform Distribution

Probability Density Function (PDF): The PDF of a continuous uniform distribution is constant across the interval ([a, b]) and is given by:
[
f(x) = \begin{cases}
\frac{1}{b-a} & \text{for } a \leq x \leq b \
0 & \text{otherwise}
\end{cases}
]



Mean: The mean (expected value) of a continuous uniform distribution is calculated as:
[
E(X) = \frac{a + b}{2}
]


Variance: The variance, which measures the spread of the distribution, is given by:
[
Var(X) = \frac{(b - a)^2}{12}
]



Applications of Continuous Uniform Distribution
Continuous uniform distributions are often used in simulations and modeling scenarios where each outcome within a range is equally likely. Common applications include:
Random number generation
Simulation of real-world processes (e.g., time until an event occurs)
Quality control in manufacturing processes




# Discrete Uniform Distribution

In contrast, a discrete uniform distribution applies to scenarios where outcomes are distinct and countable. Each outcome has an equal probability of occurring, and it is typically represented as ( U(n) ), where ( n ) is the number of possible outcomes.

Properties of Discrete Uniform Distribution







Probability Mass Function (PMF): The PMF of a discrete uniform distribution is defined as:
[
P(X = x) = \frac{1}{n} \quad \text{for } x = 1, 2, \ldots, n
]







Mean: The mean of a discrete uniform distribution can be calculated using the formula:
[
E(X) = \frac{n + 1}{2}
]







Variance: The variance for a discrete uniform distribution is given by:
[
Var(X) = \frac{(n^2 - 1)}{12}
]

Applications of Discrete Uniform Distribution



Discrete uniform distributions are commonly found in scenarios involving finite outcomes, such as:







Rolling a fair die



Drawing cards from a well-shuffled deck



Randomly selecting a participant from a group
